
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
players_DF = purchase_data.loc[:, ["Gender", "SN", "Age"]]
players_DF = players_DF.drop_duplicates()
players = players_DF.count()[0]

pd.DataFrame({"total players":[players]})

total_genders = players_DF["Gender"].value_counts()
gender_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Total purchase value")
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Total purchase value")
gender_purchase_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total purchase value")
average_purchase_price_player = gender_purchase_total/total_genders

gender_table = pd.DataFrame({"Purchase Count":gender_count, \
                             "Average Purchase Price":gender_average, \
                             "Total Purchase Value":gender_purchase_total, \
                             "Avg Total Purchase per Person":average_purchase_price_player})

gender_table["Average Purchase Price"] = gender_table["Average Purchase Price"].map("${:.2f}".format)
gender_table["Total Purchase Value"] = gender_table["Total Purchase Value"].map("${:.2f}".format)
gender_table["Avg Total Purchase per Person"] = gender_table["Avg Total Purchase per Person"].map("${:.2f}".format)

gender_table = gender_table.round(2)
gender_table
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Pop_Items = purchase_data.loc[:, ["Item ID", "Item Name", "Price" ]]
Total_Pop_Item_Count = Pop_Items["Item ID"].value_counts()
Top_Items = Total_Pop_Item_Count.loc[[178,145,108,82,19]]
Top_Items = pd.DataFrame(Top_Items)
Top_Items = Top_Items[["Item ID"]]
Top_Items = Top_Items.rename(columns={"Item ID":"Purchase Count"})
Top_Items = Top_Items[["Purchase Count"]].reset_index()
Purchase_Count = Top_Items[["Purchase Count"]]
Pop_Items_ID_Index = Pop_Items.set_index("Item ID")
Top_Items_name_price_value = Pop_Items_ID_Index.loc[[178,145,108,82,19]]
Top_Items_name_price_value = Top_Items_name_price_value.drop_duplicates()
Top_Items_name_price_value = Top_Items_name_price_value[["Item Name", "Price"]]
Top_Items_name_price_value = Top_Items_name_price_value.reset_index()
Top_Joined = Purchase_Count.join(Top_Items_name_price_value, how="outer")

Total_Purchase_Val = Top_Joined["Price"].astype(float) * Top_Joined["Purchase Count"].astype(int) 
Total_Purchase_Val = pd.DataFrame(Total_Purchase_Val)

Top_Joined_Total = Top_Joined.join(Total_Purchase_Val, how = "outer")
Top_Joined_Total.columns.values[0] = "Signal"
Top_Joined_Total
Top_Joined = Top_Joined.set_index(["Item ID", "Item Name"])
Top_Joined = Top_Joined.rename(columns={"Price":"Item Price"})
Top_Joined_Total

Top_Joined_Total_arrange = Top_Joined_Total[["Item ID","Item Name","Signal","Price",0]]
Top_Joined_Total_arrange.rename(columns={0:"Total Purchase Value"}, inplace=True)
Top_Joined_Total_arrange.rename(columns={"Signal":"Purchase Count"}, inplace=True)
Top_Joined_Total_arrange.rename(columns={"Price":"Item Price"}, inplace=True)
Top_Joined_Total_arrange = Top_Joined_Total_arrange.set_index(["Item ID", "Item Name"])

Top_Joined_Total_arrange["Item Price"] = Top_Joined_Total_arrange["Item Price"].map("${:.2f}".format)
Top_Joined_Total_arrange["Total Purchase Value"] = Top_Joined_Total_arrange["Total Purchase Value"].map("${:.2f}".format)

Top_Joined_Total_arrange
```
